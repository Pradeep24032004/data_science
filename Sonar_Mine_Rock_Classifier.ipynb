```python
#encoding the target column
sonar_dataset.iloc[:,60].replace({'M':0,'R':1}, inplace=True)
sonar_dataset
```

```python
#splitting into features and columns
X = sonar_dataset.iloc[:,0:60]
Y = sonar_dataset.iloc[:,60]
```

```python
#splitting the dataset into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
#checking the shapes of dataset, training data and testing data
print(X.shape, X_train.shape, X_test.shape)
```

```python
#calling the logistic regression model
model = LogisticRegression()
```

```python
#training the model by fitting with the training data
model.fit(X_train, Y_train)
```

```python
#model evaluation
# Import necessary libraries
from sklearn.metrics import accuracy_score
#prediction on the training data
prediction_train = model.predict(X_train)
#finding the accuracy score of the training data
accuracy_train = accuracy_score(Y_train, prediction_train)
print("Accuracy score of the training data:", accuracy_train)
```

```python
#prediction on testing data
prediction_test = model.predict(X_test)
#finding the accuracy score of the testing data
accuracy_test = accuracy_score(Y_test, prediction_test)
print("Accuracy score of the testing data:", accuracy_test)
```
