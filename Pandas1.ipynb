```python
import pandas as pd
df = pd.read_csv('nba.csv')
df
```

```python
#finding the null values in the dataset
print(df.isna()) #if there is null value then it returns true there
print('\n')
print("column wise null values \n", df.isna().sum())
```

```python
#dropping the column
df = df.drop('Height', axis = 1)
df
```

```python
#maximum salary
print("max slaary",df['Salary'].max())
print("average salary",df['Salary'].mean())
print("minimum salary",df['Salary'].min())
```

```python
#consider a player who is playing for Bosten Celtics and his position should be SF and age should be less than 22
df[(df['Team'] == 'Boston Celtics') & (df['Position'] == 'SF') &(df['Age'] > 22) ]
```

```python
#consider a player who is playing for Boston Celtics and his position should be SF,PF and Age should be greater than 22
df[(df['Team'] == 'Boston Celtics') & (df['Position'].isin(['SF','PF'])) & (df['Age'] > 22)]
```

```python
#Team = Boston Celtics and Their college must be from either Texas, Boston University find the names of the players and their salary
df[(df['Team'] == 'Boston Celtics') & (df['College'].isin(['Texas','Boston University']))][['Name','Salary']]
```

```python
#Boston Celtics, Los Angeles Lakers, Houston Rockets, Denver Nuggets, Utah Jazz how many players are there in these teams
df['Team'].isin(['Boston Celtics','Los Angeles Lakers','Houston Rockets','Denver Nuggets','Utah Jazz']).value_counts()
```

```python
#Boston Celtics, Los Angeles Lakers, Houston Rockets, Denver Nuggets, Utah Jazz and position in sf, sg
df[(df['Team'].isin(['Boston Celtics','Los Angeles Lakers','Houston Rockets','Denver Nuggets','Utah Jazz'])) & (df['Position'].isin(['SF','SG'])) ]
```

```python
#Boston Celtics, Los Angeles Lakers, Houston Rockets, Denver Nuggets, Utah Jazz and position in sf, sg and salary should be greater than 70,00,000
df[(df['Team'].isin(['Boston Celtics','Los Angeles Lakers','Houston Rockets','Denver Nuggets','Utah Jazz'])) & (df['Position'].isin(['SF','SG'])) & (df['Salary'] > 7000000) ]
```

```python
#Boston Celtics, Los Angeles Lakers, Houston Rockets, Denver Nuggets, Utah Jazz or position in sf, sg
df[(df['Team'].isin(['Boston Celtics','Los Angeles Lakers','Houston Rockets','Denver Nuggets','Utah Jazz'])) | (df['Position'].isin(['SF','SG'])) & (df['Salary'] > 7000000) ]
```
