```python
#creating a series from a List
import pandas as pd
data = [1,2,3,4,5]
series_from_list = pd.Series(data)
print("series from list",series_from_list)

#creating a series with custom index using dictionary
data1 = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}
series_from_dict = pd.Series(data1)
print("series from dictionary",series_from_dict)
```

```python
#Data Frame
'''
In the real world, a pandas DataFrame will be created by loading the datasets from existing storage, storage can be SQL Database,
csv file, or an Exel File. Pandas DatFrame can be created from lists, dictionaries, and from a list of dictionaries, etc.
'''
import pandas as pd
df = pd.DataFrame()
df
```

```python
#Example 1 : Creating a data frame using lists
data = [
    ['Alice',25,'New York'],
    ['Bob',30,'San Francisco'],
    ['Charlie',35,'Los Angels']
]
#custom column creation
custom_columns = ['Name','Age','City']
df = pd.DataFrame(data,columns=custom_columns)
print(df)
```

```python
#Example 2 : creating a data frame using dictionary
data1 = {
    'Name':['Alice','Bob','Charlie'],
    'Age':[25,30,35],
    'City':['New York','San Francisco','Los Angeles']
}
df = pd.DataFrame(data1)
print(df)
```

```python
'''
Analyzing a DataFrame

Create a dataframe with information about sales data for a small store.
The DataFrame should include columns for "Product Name", "Price" and "Quality Sold".
Then, calculate the total revenue for each product(price * Quantity Sold) and display the DataFrame with the total revenue column added.
'''
```

```python
import pandas as pd
data = {
    'Product Name':['Product A','Product B','Product C','Product D'],
    'Price':[10.99,5.99,8.49,12.49],
    'Quantity Sold':[100,150,80,200]
}
#step 1: converting dictionary into a dataframe
sales_df = pd.DataFrame(data)
print(sales_df)
print('\n')
#step 2: calculating the total revenue
sales_df['total revenue'] = sales_df['Price']*sales_df['Quantity Sold']
print(sales_df)
sales_df['Discount'] = sales_df['Price']*0.1
print('\n')
print(sales_df)
print('\n')
sales_df['Revised_product_price_after_discount']=sales_df['Price'] - sales_df['Discount']
print(sales_df)
```

```python
#dropping the missing values from the dataset
print(df.dropna()) #dropping all the null values
print('\n')
#dropna using how parameter
print("using how=any \n",df.dropna(how='any')) #if any column in aparticular row is null then it drops the entire row
print("using how=all \n",df.dropna(how='all')) #drops all null valued columns
```

```python
#to permunantly manupilate the dataset
df.dropna(how='all',inplace=True) #makes permenant changes, drops all the null values
print(df)
```
