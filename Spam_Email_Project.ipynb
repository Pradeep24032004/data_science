```python
raw_mail_data = pd.read_csv('mail_data.csv')
raw_mail_data #raw_mail_data is data frame
```

```python
#replacing all the null values with null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
#printing the first five rows
mail_data.head()
```

```python
#checking the rows and columns of the dataset
mail_data.shape
```

```python
#Label encoding: converting spam into 0 and ham into 1
mail_data.loc[mail_data['Category']=='spam', 'Category'] = 0
mail_data.loc[mail_data['Category']=='ham', 'Category'] = 1
mail_data
```

```python
#separating the data as text and label
X = mail_data['Message']
Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting the data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
#transforming the text data to feature vectors that can be used as input to the logistic regression
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english')
X_train_num = feature_extraction.fit_transform(X_train)
X_test_num = feature_extraction.transform(X_test) #it will just transform the dataset
```
